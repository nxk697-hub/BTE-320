```python
# Inputs: sales ($) and cost of goods pct (%)
sales = input("Enter sales ($): ")
sales = float(sales)

cogs_pct = input("Enter Cost of Goods (%): ")
cogs_pct = float(cogs_pct)

rate = input("Enter tax rate (%): ")
rate = float(rate)

# Processes: calculate cost of goods sold and gross profit
cogs = sales * (cogs_pct / 100)
grossProfit = sales - cogs
taxes = grossProfit * rate / 100
netIncome = grossProfit - taxes

# Outputs: display cost of goods sold and gross profit
print(f"Cost of Goods sold: ${cogs}")
print(f"Gross Profit: ${grossProfit}")
print(f"Taxes: ${taxes}")
print(f"Net income: ${netIncome}")
```

```python
import random

play = input('Do you want to play? ').strip().lower()

while play=='yes' and play!='no':
  computer = random.choice(['rock', 'paper', 'scissors'])
  user = input("User choice (rock, paper, scissors): ").strip().lower()

  if user == computer:
    print(f'User:{user}, Computer:{computer}')
    print('Draw')
  elif (
    (user == 'rock' and computer == 'scissors') or
    (user == 'scissors' and computer == 'paper') or
    (user == 'paper' and computer == 'rock')):
    print(f'User:{user}, Computer:{computer}')
    print('You win!')
  else:
    print(f'User:{user}, Computer:{computer}')
    print('Computer wins')

  play=input("Do you wish to continue?")
```

```python
def getValues(n: int) -> list:
  '''
  Creates a list of weights. The weights are given by the user
  Input:
  - number of item in the list
  Output:
  - a list of weights
  '''
  items = []
  for i in range(n):
    try:
      item = float(input('Enter weight value:'))
    except Exception as e:
      print(e)
      item = float('nan')

    items.append(item)

  return items

def calcAverage(alist: list) -> float:
  '''
  Input:
  - A list of weights
  Output:
  - Average value; excludes the first 2 items
  '''
  try:
    return sum(alist[2:])/len(alist[2:])
  except ZeroDivisionError:
    return float('nan')

def main():
  num = int(input('Enter number of weights:'))

  if num <= 0:
    raise ValueError('num must be positive')

  weights = getValues(num)
  avg = calcAverage(weights)
  print(f'Average weight: {avg} lbs.')

main()
```

```python
class Greeting:
  def __init__(self,first, last):
    self.first=first
    self.last=last

  def __str__(self):
    return "Instance of class Greeting"

  def display(self, reverse=True):
    if reverse:
      print(f'{self.last}, {self.first}')
    else:
      print(self.first)

name = Greeting('John', 'Doe')
print(name)
name.display()
name.display(reverse=False)

```
